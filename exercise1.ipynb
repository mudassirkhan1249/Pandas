How many unique countries are there in the dataset?

```python
df['country'].unique()
df['country'].nunique()
```

How many continents are covered?

```python
usingVCount = df['continent'].value_counts()
usingUnique = df['continent'].nunique()
```

Which continent has the highest number of countries?

```python
df['continent'].value_counts()
```

```python
# Only continent name
df['continent'].value_counts().idxmax()
```

Calculate the average population for each continent.

```python
df.groupby('continent')['population'].mean()
```

```python
print(f"{df.groupby('continent')['population'].mean().idxmax()} = {df.groupby('continent')['population'].mean().max()}")
```

List the top 10 most populated countries.

```python
top10 = df[['country','population']].sort_values(by='population', ascending = False).head(10)
top10.reset_index(drop=True)
```

Which country has the highest median age?

```python
df[['country','median_age']].sort_values(by='median_age',ascending=False).head().reset_index(drop=True)
```

Which region has the highest rural population (%)?

```python
df[['country','rural_population']].sort_values(by='rural_population',ascending=False).head().reset_index(drop=True)
```

Which continent has the highest average democracy score?

```python
df.groupby('continent')['democracy_score'].mean().idxmax()
```

How many countries have a missing capital city?

```python
df['capital_city'].isna().sum()
```

Which country has the best press freedom score (highest or lowest ‚Äî according to how the dataset defines it)?

```python
df[['country','press']].sort_values(by='press',ascending=False).head(1).reset_index(drop=True)
```
