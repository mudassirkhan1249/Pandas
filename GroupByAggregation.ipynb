```python
import pandas as pd
import numpy as np
```

# Group By

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

Category wise total sales

```python
category = df.groupby("Category")["Sales"].sum()
category
```

Store wise average sales

```python
store = df.groupby("Store")["Sales"].mean()
store
```

Category + Store wise total quantity

```python
cat_store_qty = df.groupby(['Category', 'Store'])["Quantity"].sum()
cat_store_qty
```

group by store and sum of all Sales

```python
st_sales = df.groupby("Store")["Sales"].sum()
st_sales
```

## Aggregation

```python
df["Sales"].agg(["mean","min","max","std","median","count"])
```

```python
df.groupby("Category")["Sales"].agg(["mean","max","min","count","std","median"])
```
