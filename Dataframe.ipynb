# Data Frame

## Creating DataFrame

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 101200, 85000]
}
```

```python
df = pd.DataFrame(data)
df
```

## Selecting and Indexing of columns

```python
df
```

Selecting one column

```python
df["Name"]
```

Selecting multiple column

```python
df[["Name", "Salary", "City"]]
```

## Creating new column

```python
df["Designation"] = ["CA", "Software Engineer", "AI Architecture", "CEO"]
```

```python
df
```

## Remove column

Remove a single column 

```python
# Temporary removing

df.drop("Designation", axis = 1)
```

```python
df
```

```python
# Permanent removing

df.drop("Designation", axis = 1, inplace = True)
```

```python
df
```

Removes multiple column 

```python
df.drop(["City", "Age"], axis = 1)
```

```python
df
```

## Conditional Selection

```python
df
```

```python
df[df["Salary"] > 80000]
```

```python
# AND
df[(df["Salary"] > 80000) & (df["Age"] > 28)]
```

```python
# OR
df[(df["Salary"] > 80000) | (df["Age"] > 40)]
```
