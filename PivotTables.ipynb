```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Date": pd.date_range(start="2023-01-01", periods=30),
    
    "Product": np.random.choice(["Laptop", "Mobile", "Tablet", "Headphones"], 30),
    
    "Region": np.random.choice(["North", "South", "East", "West"], 30),
    
    "Sales": np.random.randint(5000, 50000, 30),
    
    "Units": np.random.randint(1, 20, 30),
    
    "Sales_Person": np.random.choice(
        ["Ali", "Ahmed", "Sara", "Ayesha", "Bilal"], 30
    ),
}

df = pd.DataFrame(data)

# Extra useful columns
df["Month"] = df["Date"].dt.month_name()
df["Quarter"] = "Q" + df["Date"].dt.quarter.astype(str)

df
```

# Pivot Table

```python
pd.pivot_table(df, values = "Sales", index = "Region", columns = "Product")
```

```python
pd.pivot_table(df, values = "Sales", index = "Region", columns = "Product",aggfunc = "count")
```

```python
pd.pivot_table(df, values = ["Sales", "Units"], index = "Product", columns = "Sales_Person")
```

```python
pivot2 = pd.pivot_table(df, values = "Sales", index = "Product", columns = "Sales_Person", aggfunc = "sum")
pivot2.fillna(0)
```

# Cross Tab

```python
pd.crosstab(df["Region"],df["Product"])
```
